```python
import numpy as np
from stl import mesh
```

```python
vertices = np.array([
    [0, 0, 0],    # Vertex 0
    [17, 0, 0],   # Vertex 1
    [17, 12, 0],  # Vertex 2
    [0, 12, 0],   # Vertex 3
    [0, 0, 1],    # Vertex 4
    [17, 0, 1],   # Vertex 5
    [17, 12, 1],  # Vertex 6
    [0, 12, 1],   # Vertex 7
    [1, 1, 1],    # Vertex 8
    [1, 11, 1],   # Vertex 9
    [16, 1, 1],   # Vertex 10
    [16, 11, 1],  # Vertex 11
    [1, 1, 0.5],  # Vertex 12
    [1, 11, 0.5], # Vertex 13
    [16, 1, 0.5], # Vertex 14
    [16, 11, 0.5] # Vertex 15
])

faces = np.array([
    [0, 3, 1], [1, 3, 2],       # Bottom face
    [0, 4, 7], [0, 7, 3],       # Left face
    [5, 1, 2], [5, 2, 6],       # Right face
    [2, 3, 6], [3, 7, 6],       # Front face
    [0, 1, 5], [0, 5, 4],       # Back face

    [12, 13, 14], [13, 14, 15], # Middle bottom face
    [12, 8, 13], [8, 13, 9],    # Middle left face
    [13, 9, 15], [9, 15, 11],   # Middle front face
    [12, 8, 14], [8, 14, 10],   # Middle back face
    [14, 10, 15], [10, 15, 11], # Middle right face

    [4, 7, 8], [7, 8, 9],       # Top left face
    [5, 6, 10], [6, 11, 10],    # Top right face
    [7, 9, 6], [6, 9, 11],      # Top front face
    [4, 8, 5], [5, 8, 10]       # Top back face
])
```

```python
frame = mesh.Mesh(np.zeros(faces.shape[0], dtype=mesh.Mesh.dtype))
for i, f in enumerate(faces):
    for j in range(3):
        frame.vectors[i][j] = vertices[f[j],:]

frame.save('Frame.stl')
```
